## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal'] = df['unit_price']  * df['qty']
columns_to_display = ['unit_price','qty', 'line_subtotal']
df[columns_to_display].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']

df['shipping_price'] = df.apply(lambda row: 7 * row['total_weight'] if row ['total_weight'] > 50 else 10 * row['total_weight'], axis=1)
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df.apply(lambda row: 7 * row['total_weight'] if row['total_weight'] > 50 else 10 * row['total_weight'], axis=1)
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()

```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['line_price'] = (df['line_subtotal'] + df['shipping_price'] + (df['line_subtotal'] + df['shipping_price']) * 0.0925).round(2)
df[['line_subtotal', 'shipping_price', 'line_price']].head()
df['line_profit'] = df['line_price'] - df['line_cost']
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
desired_columns = [
        'first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_date', 'order_week', 'order_year', 'unit_price', 'unit_cost', 'unit_weight', 'qty', 'line_number', 'line_subtotal', 'total_weight', 'shipping_price', 'line_price', 'line_cost', 'line_profit'
]
df[desired_columns].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_clients = df['client_id'].value_counts().head(5).index.tolist()
top_clients_df = df[df['client_id'].isin(top_clients)]
total_spending_by_client = top_clients_df.groupby('client_id')['line_price'].sum()
total_spending_by_client.index.name = None
total_spending_by_client.apply(lambda x: f"${x:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

top_clients = df['client_id'].value_counts().head(5).index.tolist()
top_clients_df = df[df['client_id'].isin(top_clients)]

client_summary = top_clients_df.groupby('client_id').agg(
    qty=('qty', 'sum'),
    shipping_price=('shipping_price', 'sum'),
    line_price=('line_price', 'sum'),
    line_cost=('line_cost', 'sum'),
    line_profit=('line_profit', 'sum')
)
client_summary = client_summary.sort_values(by='line_profit', ascending=False)
client_summary
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

client_summary_format = client_summary.copy()
client_summary_format['shipping_price'] /= 1e6
client_summary_format['line_price'] /= 1e6
client_summary_format['line_cost'] /= 1e6
client_summary_format['line_profit'] /= 1e6

client_summary_format = client_summary_format.rename(
    columns={
        'qty': 'Units',
        'shipping_price': 'Shipping',
        'line_price': 'Total Revenue',
        'line_cost': 'Total Cost',
        'line_profit': 'Total Profit'
    }
)

client_summary_format['Shipping'] = client_summary_format['Shipping'].apply(lambda x: f"${x:.2f}M")
client_summary_format['Total Revenue'] = client_summary_format['Total Revenue'].apply(lambda x: f"${x:.2f}M")
client_summary_format['Total Cost'] = client_summary_format['Total Cost'].apply(lambda x: f"${x:.2f}M")
client_summary_format['Total Profit'] = client_summary_format['Total Profit'].apply(lambda x: f"${x:.2f}M")


client_summary_format
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
client_summary_format = client_summary_format.sort_values(by='Total Profit', ascending=False)
client_summary_format
```
